## **Ristentroopia kõverad**

```python
## Data
y_true = [0, 0, 0, 1, 1]
p_pred = [0.1, 0.2, 0.3, 0.75, 0.8]

## Plot figure 
plt.figure(figsize=(8,5))

def cross_entropy(c):
    if c == 0 or c==1:
        return 1000
    return 1 / (c * (1-c))

## Plot curve for all the data.
y = brier_curve(y_true, p_pred, w=cross_entropy)
plt.plot(x, y, label='whole curve', linewidth=3)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.xlabel('Hindade suhe C', fontsize=16)
plt.ylabel('1\(c(1-c)) * Keskmine hind',  fontsize=16)

plt.savefig('photos/crossentropy_distorted_curve.png')
```
